AIM #2: Finding issues (empty, NAs, incorrect value, incorrect format, outliers, etc.) 
1. Find out how many missing values there are in the dataset
2. For the 'Age' column, find the best way to handle the missing values
    2.1. Use an appropriate plot to study the nature of the 'Age' column
    2.2. Figure out what is the best way to calculate the central tendency of the 'Age' column based on the above plot
    2.3. Using the most suitable central tendency measure, fill the missing values in the age column

```python
import pandas as pd 
import matplotlib.pyplot as plt 


df = pd.read_csv('titanic.csv') 


missing_values = df.isnull().sum()
print("\nMissing Values in Each Column:")
print(missing_values)


plt.figure(figsize=(10, 6))
plt.hist(df['Age'].dropna(), bins=30, edgecolor='black', alpha=0.7)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()


mean_age = df['Age'].mean()
median_age = df['Age'].median()
print(f"Mean Age: {mean_age}, Median Age: {median_age}")


df['Age'].fillna(median_age, inplace=True)


df['Cabin'].fillna('Unknown', inplace=True)


mode_embarked = df['Embarked'].mode()[0]
df['Embarked'].fillna(mode_embarked, inplace=True)


print("\nUnique values in 'Survived':")
print(df['Survived'].unique())


df['Fare'] = pd.to_numeric(df['Fare'], errors='coerce')


df['Fare'].fillna(df['Fare'].median(), inplace=True)


print("\nMissing Values After Cleaning:")
print(df.isnull().sum())
```

AIM #3: Grouping 
1. Find out the average fare grouped by Pclass
    1.1. Plot the above using a suitable plot
2. Find out the average fare grouped by Sex
    2.1. Plot the above using a suitable plot

```python
import pandas as pd 
import matplotlib.pyplot as plt 


df = pd.read_csv('titanic.csv') 

df['Fare'] = pd.to_numeric(df['Fare'], errors='coerce')
df['Fare'].fillna(df['Fare'].median(), inplace=True)

average_fare_pclass = df.groupby('Pclass')['Fare'].mean()
print("\nAverage Fare by Pclass:")
print(average_fare_pclass)


plt.figure(figsize=(10, 6))
average_fare_pclass.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Fare by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Average Fare')
plt.xticks(rotation=0)
plt.show()


average_fare_sex = df.groupby('Sex')['Fare'].mean()
print("\nAverage Fare by Sex:")
print(average_fare_sex)


plt.figure(figsize=(10, 6))
average_fare_sex.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Average Fare by Sex')
plt.xlabel('Sex')
plt.ylabel('Average Fare')
plt.xticks(rotation=0)
plt.show()
```

AIM #4: Dataset visualization using pandas

1. Plot the distribution of 'Age' using a suitable plot
2. Plot the distribution of 'Fare' using a suitable plot
3. Plot the distribution of 'Pclass' using a suitable plot
4. Plot the distribution of 'Survived' using a suitable plot
5. Plot the distribution of 'Embarked' using a suitable plot
6. Plot the distribution of 'Fare' grouped by 'Survived'
7. Plot the distribution of 'Fare' grouped by 'Pclass'
8. Plot the distribution of 'Age' grouped by 'Survived'
9. Plot the distribution of 'Age' grouped by 'PClass'
10. Combine the 'SibSp' and 'Parch' and plot its distribution grouped by 'Survived'
11. Combine the 'SibSp' and 'Parch' and plot its distribution grouped by 'Pclass'
12. Plot a distribution between 'Age' and 'Fare' to see if there's any relationship
13. Are there any other possibilities to show relationships?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Titanic dataset
df = pd.read_csv('titanic.csv')

# 数据清洗
df['Fare'] = pd.to_numeric(df['Fare'], errors='coerce')
df['Fare'].fillna(df['Fare'].median(), inplace=True)
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Cabin'].fillna('Unknown', inplace=True)
mode_embarked = df['Embarked'].mode()[0]
df['Embarked'].fillna(mode_embarked, inplace=True)
df['Survived'] = pd.to_numeric(df['Survived'], errors='coerce').fillna(0).astype(int)

# 1. 绘制年龄分布（核密度估计图 KDE）
plt.figure(figsize=(10, 6))
df['Age'].plot(kind='kde', color='blue')
plt.title('Age Distribution (KDE Plot)')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()

# 2. 票价分布（累计分布函数 CDF）
plt.figure(figsize=(10, 6))
df['Fare'].plot(kind='hist', bins=50, density=True, cumulative=True, color='green', edgecolor='black', alpha=0.7)
plt.title('Fare Distribution (CDF)')
plt.xlabel('Fare')
plt.ylabel('Cumulative Density')
plt.show()

# 3. Pclass 分布（堆叠条形图）
plt.figure(figsize=(10, 6))
pclass_survived = pd.crosstab(df['Pclass'], df['Survived'])
pclass_survived.plot(kind='bar', stacked=True, color=['red', 'green'], edgecolor='black')
plt.title('Pclass Distribution (Stacked Bar Chart)')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

# 4. 生存情况分布（条形图）
plt.figure(figsize=(10, 6))
df['Survived'].value_counts().plot(kind='bar', color=['lightcoral', 'lightgreen'], edgecolor='black')
plt.title('Survived Distribution (Bar Chart)')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

# 5. 登船港口分布（堆叠条形图按生存情况分组）
plt.figure(figsize=(10, 6))
embarked_survived = pd.crosstab(df['Embarked'], df['Survived'])
embarked_survived.plot(kind='bar', stacked=True, color=['skyblue', 'orange'], edgecolor='black')
plt.title('Embarked Distribution by Survived (Stacked Bar Chart)')
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

# 6. 按生存情况绘制票价分布（小提琴图）
plt.figure(figsize=(10, 6))
df.boxplot(column='Fare', by='Survived', grid=False)
plt.title('Fare Distribution by Survived (Violin Plot)')
plt.suptitle('')
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.show()

# 7. 按Pclass绘制票价分布（小提琴图）
plt.figure(figsize=(10, 6))
df.boxplot(column='Fare', by='Pclass', grid=False)
plt.title('Fare Distribution by Pclass (Violin Plot)')
plt.suptitle('')
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.show()

# 8. 按生存情况绘制年龄分布（核密度估计图 KDE）
plt.figure(figsize=(10, 6))
df.groupby('Survived')['Age'].plot(kind='kde', legend=True)
plt.title('Age Distribution by Survived (KDE Plot)')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()

# 9. 按Pclass绘制年龄分布（核密度估计图 KDE）
plt.figure(figsize=(10, 6))
df.groupby('Pclass')['Age'].plot(kind='kde', legend=True)
plt.title('Age Distribution by Pclass (KDE Plot)')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()

# 10. 合并SibSp和Parch，按生存情况绘制分布（散点图）
df['FamilySize'] = df['SibSp'] + df['Parch']
plt.figure(figsize=(10, 6))
plt.scatter(df['Survived'], df['FamilySize'], color='purple', s=100, alpha=0.5)
plt.title('Average Family Size by Survived (Scatter Plot)')
plt.xlabel('Survived')
plt.ylabel('Average Family Size')
plt.xticks([0, 1], ['Did Not Survive', 'Survived'], rotation=0)
plt.show()

# 11. 合并SibSp和Parch，按Pclass绘制分布（散点图）
plt.figure(figsize=(10, 6))
plt.scatter(df['Pclass'], df['FamilySize'], color='cyan', s=100, alpha=0.5)
plt.title('Average Family Size by Pclass (Scatter Plot)')
plt.xlabel('Pclass')
plt.ylabel('Average Family Size')
plt.xticks(rotation=0)
plt.show()

# 12. 绘制年龄与票价之间的关系（气泡图）
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Fare'], alpha=0.5, s=df['FamilySize']*10, color='green', edgecolor='black')
plt.title('Age vs Fare (Bubble Chart)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()

# 13. 其他关系的可视化：年龄与生存情况的关系（密度图）
plt.figure(figsize=(10, 6))
df.groupby('Survived')['Age'].plot(kind='kde', legend=True)
plt.title('Age Distribution by Survived (Density Plot)')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

AIM #5: Correlation

1. Generate a correlation matrix for the entire dataset
2. Find correlation between 'Age' and 'Fare'
3. What other possible correlations can be found in the dataset?

```python
import pandas as pd 
import matplotlib.pyplot as plt 

# 读取数据
df = pd.read_csv('titanic.csv')

# 检查并处理缺失值
df['Fare'] = pd.to_numeric(df['Fare'], errors='coerce')
df['Fare'].fillna(df['Fare'].median(), inplace=True)
df['Age'].fillna(df['Age'].median(), inplace=True)
mode_embarked = df['Embarked'].mode()[0]
df['Embarked'].fillna(mode_embarked, inplace=True)

# 确保Survived列为整数类型
df['Survived'] = pd.to_numeric(df['Survived'], errors='coerce').fillna(0).astype(int)

# 处理其他非数值列
# 将 'Sex' 列转换为数值（0和1）
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})

# 删除非数值列
df = df.select_dtypes(include=['float64', 'int64'])

# 1. 生成相关性矩阵
correlation_matrix = df.corr()
print("Correlation Matrix:")
print(correlation_matrix)

# 2. 计算年龄与票价之间的相关性
age_fare_correlation = df['Age'].corr(df['Fare'])
print(f"\nCorrelation between Age and Fare: {age_fare_correlation}")

# 3. 查找其他可能的相关性
strong_correlations = correlation_matrix[(correlation_matrix.abs() > 0.3) & (correlation_matrix != 1)]
print("\nStrong correlations (|correlation| > 0.3):")
print(strong_correlations)

# 可视化相关性矩阵
plt.figure(figsize=(10, 8))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.title('Correlation Heatmap')
plt.show()
```
